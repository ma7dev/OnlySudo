### Load sample audio file

```python
from IPython.display import Audio 
Audio('samples/demo.wav', autoplay=False)
```

### Classification model

```python
from klaam import SpeechClassification
model = SpeechClassification()
model.classify('samples/demo.wav')
```

### Recongition model

```python
 from klaam import SpeechRecognition
 model = SpeechRecognition()
 model.transcribe('samples/demo.wav')
```

### Text to Speech

```python
from klaam import TextToSpeech
from IPython.display import Audio

model = TextToSpeech()
model.synthesize("وَسادْيُو مَانِي فِي الدَّقِيقَتَيْنِ السّابِعَةَ عَشْرَةَ - وَ الْخَامِسَةَ وَالْأَرْبعينَ مِنَ الشَّوْطِ الْأَوَّلِ لِلْمُبَارَاةِ")
Audio("sample.wav")
```
